```python
# Write your code here
!pip install tweepy
#import the library used to query a website
import json
import csv
import tweepy
import re

def hashtag_search(consumer_key, consumer_secret, access_token, access_token_secret, hashtag):
  
  # create authentication for accessing Twitter
  auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_token, access_token_secret)

  # initialize Tweepy API
  api = tweepy.API(auth, wait_on_rate_limit=True)

  # get the name of the spreadsheet we will write to
  fname = '_'.join(re.findall(r"#(\w+)", hashtag))

  # will collect 1000 posts
  numtweet = int(input('Enter the number of tweets that you want to collect: '))

  # open the spreadsheet we will write to
  with open('%s.csv' % (fname), 'w') as file:
      w = csv.writer(file)
      
      # write header row to spreadsheet
      w.writerow(['posted time', 'tweet text', 'user_name', 'location'])

      # for each tweet matching our hashtags, write relevant info to the spreadsheet
      for tweet in tweepy.Cursor(api.search, q=hashtag+' -filter:retweets', lang="en", tweet_mode='extended').items(numtweet):
            w.writerow([tweet.created_at, tweet.full_text.replace('\n',' ').encode('utf-8'), tweet.user.screen_name.encode('utf-8'), tweet.user.location.encode('utf-8')])

consumer_key = input('Consumer Key:  ')
consumer_secret= input('Consumer Secret: ')
access_token = input('Access Token: ')
access_token_secret = input('Access Token Secret: ')

hashtag = input("Enter the hashtag you want to scrape: ")

if __name__ == '__main__':
    hashtag_search(consumer_key, consumer_secret, access_token, access_token_secret, hashtag)

    
```

```python
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
# add appropriate words that will be ignored in the analysis
ADDITIONAL_STOPWORDS = ['covfefe']

def basic_clean(text):
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english') + ADDITIONAL_STOPWORDS
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

**Write your explanations of the constituency parsing tree and dependency parsing tree here (Question 3-2):** 

- Constituency parsing is the process of analyzing the sentences by breaking down it into sub-phrases. The sub-phrases are specific category of grammar like NP (noun phrase) and VP (verb phrase)
- Dependency parsing is the process of analyzing the grammatical structure of a sentence on the dependencies between the words in a sentence.
- Example: "The world is one big data problem."
  + Constituency parsing: first level is sentence, which is splitted into NP ("The world") and VP ("is one big data problem."). The VP is divided into a VP and another NP. In this example, VP is "is" and NP is "one big data problem." And continuosly until it could not be broken anymore.
  + Dependency parsing: it is completely different. The root is the verb of sentence ("is"). "world" is nsubj, "one" is nummod, "big" is amod, "data" dobj, "problem" is dobj and "The" is determiner.
